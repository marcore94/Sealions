```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Read metrics from file

metrics_unweighted_df = pd.read_csv('./binary_classifier/metrics_lr0,0005_epochs60.csv',
                                    dtype={"acc": float,
                                           "loss": float,
                                           "val_acc": float,
                                           "val_loss": float})
metrics_unweighted = metrics_unweighted_df.as_matrix()

metrics_weighted_df = pd.read_csv('./binary_classifier/metrics_lr0,0005_epochs60_weight4-1.csv',
                                    dtype={"acc": float,
                                           "loss": float,
                                           "val_acc": float,
                                           "val_loss": float})
metrics_weighted = metrics_weighted_df.as_matrix()
```

```python
# Summarize history for accuracy
plt.plot(metrics_unweighted[:, 0])
plt.plot(metrics_unweighted[:, 2])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'validation'], loc='upper left')
plt.show()
# Summarize history for loss
plt.plot(metrics_unweighted[:, 1])
plt.plot(metrics_unweighted[:, 3])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'validation'], loc='upper left')
plt.show()
```

```python
# Summarize history for accuracy
plt.plot(metrics_weighted[:, 0])
plt.plot(metrics_weighted[:, 2])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'validation'], loc='upper left')
plt.show()
# Summarize history for loss
plt.plot(metrics_weighted[:, 1])
plt.plot(metrics_weighted[:, 3])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'validation'], loc='upper left')
plt.show()
```
